```python
import pandas as pd
```

```python
url = "https://docs.google.com/spreadsheets/d/1bTNuLS7_URVMA5JGOF6fYngXyRPa2ylnJHJ7HCjEYvo/export/format=xlsx"


file = pd.ExcelFile(url)
```

```python
print(file)
```

```python
file.sheet_names
```

```python
# creating empty list of dataframes

dfs = []


# iterating over each sheet
for sheet in file.sheet_names:
    df = file.parse(sheet)
    dfs.append(df)
```

```python
lego_df = pd.concat(dfs, ignore_index=True)
```

```python
lego_df
```

```python
lego_df = lego_df.drop(columns ="id")
```

```python
lego_df
```

```python
list_columns = list(lego_df.columns)
```

```python
for column in list_columns:
    print(lego_df[column].nunique())
```

```python
for column in list_columns:
    print(f"the number of unique values in this {column} is {lego_df[column].nunique()}")
```
