## Renaming columns

```python
products_df = products_df.rename(columns={'Country (where company is incorporated)': 'country'})
products_df = products_df.rename(columns={"*Company's sector": 'sector'})
```

## How many countries do we have?

```python
products_df["country"].value_counts()
```

## Creating boxplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(data=products_df, x="sector", y="*Carbon intensity", log_scale=True)
plt.xticks(rotation=90)  # For x-axis labels
plt.tick_params(axis='both', which='major', labelsize=8)
#plt.yticks(rotation=angle)  # For y-axis labels
plt.show()
```

## Creating Cola Cola dataframe

```python
cola_df = products_df.loc[products_df['Company'].str.contains('cola', case=False)]
cola_df = cola_df.reset_index(drop=True)
cola_df
```

```python
cola_df.columns
```

## Merging dataframes

```python
merged_df = pd.merge(cola_df, stages_df, on='*PCF-ID', how='inner')
```

```python
merged_df
```

```python
cola_stage_df =merged_df[["Product name (and functional unit)", 
                          "Description of LCA stage",
                          "*Assigned value chain portion",
                          "Emissions at stage (kg CO2e)"]]
cola_stage_df
```
